```python
#histogram of insulin
plt.figure(figsize=(5,5))
plt.hist(df['Insulin'],bins=10,color='green')
plt.xlabel('Insulin')
plt.ylabel('Frequency')
plt.title('histogram of insulin')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.hist(df['Age'],bins=10,color='green')
plt.xlabel('Age')
plt.ylabel('Age')
plt.title('histogram of Age')
plt.show()
```

```python
#boxplot of Age
plt.figure(figsize=(10,5))
plt.boxplot(df['Age'])
plt.xlabel('Age')
plt.ylabel('Age')
plt.title('boxplot of age')
plt.show()
```

```python
#barchart of glucose and outcome
plt.figure(figsize=(5,5))
plt.bar(df['Glucose'],df['Outcome'])
plt.xlabel('Glucose')
plt.ylabel('Outcome')
plt.title('barchart of outcome and glucose')
plt.show()
```

```python
#scatter plot of glucose levels vs. blood pressure
plt.figure(figsize=(8,6))
plt.scatter(df['Glucose'],df['BloodPressure'])
plt.xlabel('Glucose')
plt.ylabel('BloodPressure')
plt.title('Scatter plot of Glucose vs. Blood Pressure')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Glucose'],df['BloodPressure'], c=df['Outcome'], cmap='PuBu')
plt.xlabel('Glucose')
plt.ylabel('BloodPressure')
plt.title('Scatter plot of Glucose vs. Blood Pressure')

#Add colorbar and legend
colorbar = plt.colorbar(label='Outcome')
colorbar.set_ticks([0,1])
colorbar.set_ticklabels(['Non-Diabetic', 'Diabetic'])
```

```python
plt.figure(figsize=(8,6))
plt.pie(df['Outcome'].value_counts(), labels=['Non-Diabetic', 'Diabetic'], autopct='%1.1f%%')
plt.title('Pie Chart of Outcome')
plt.show()   
```
